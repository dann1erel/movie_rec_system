# Функции для работы

```python
def get_genres(df):
    genres_raw = df['genres'].unique().tolist()
    genres = []
        
    for genre in genres_raw:
        if ', ' in genre:
            genres.extend(genre.split(', '))
        else:
            genres.append(genre)

    genres = set(genres)
    i = 0
    d = {genre: 0 for genre in genres}

    for item in df['genres']:
        for genre in genres:
            if genre in item:
                d[genre] += 1
    d = list(dict(sorted(d.items(), key=lambda item: item[1], reverse=True)[:20]).keys())
    return d


def get_films_with_genres(df):
    genres = get_genres(df)
    pattern = '|'.join(genres)
    return df[df['genres'].str.contains(pattern, regex=True)]
```

# Работа с датасетом Кинопоиска (kp) с нужными жанрами

```python
kp.head()
```

```python
kp = kp.dropna(subset=['genres', 'countries'])

f = lambda x: x[1:-1]
kp['genres'] = kp['genres'].apply(f)
kp['countries'] = kp['countries'].apply(f)

kp = get_films_with_genres(kp)
kp = kp[(kp['movie_year'] > 1990) & (kp['kp_rating_count'] > kp['kp_rating_count'].median())]
kp.to_csv('../data/kp_new_genres.csv')
```
