# Получаем список жанров из датасета онлайн-кинотеатр kion

```python
kion = pd.read_csv('../data/kion_data.csv', index_col='item_id')
kion.sample(10)
```

```python
genres_raw = kion.genres.unique().tolist()
genres_raw
```

```python
genres = []

for genre in genres_raw:
    if ', ' in genre:
        genres.extend(genre.split(', '))
    else:
        genres.append(genre)

genres = set(genres)
genres, len(genres)
```

```python
i = 0
d = {genre: 0 for genre in genres}

for item in kion.genres:
    for genre in genres:
        if genre in item:
            d[genre] += 1

d
```

```python
d = list(dict(sorted(d.items(), key=lambda item: item[1], reverse=True)[:20]).keys())
d
```
